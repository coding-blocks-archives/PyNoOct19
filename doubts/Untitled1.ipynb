```python
url = "https://graph.facebook.com/{}/picture?type=large"
```

```python
for i in range(4, 200):
    r = requests.get(url.format(i))
    with open("images/{}.png".format(i), "wb") as file:
        file.write(r.content)
    print("done with", i)
```
